```python
from qiskit import QuantumCircuit
from qiskit.converters import circuit_to_gate
from qiskit.visualization import array_to_latex
from qiskit.quantum_info import Operator
from qiskit.quantum_info import Statevector
from qiskit import transpile 
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram
from qiskit.circuit import ControlledGate
import math 

# The aim of the assignment is to simulate the Ekert91 key distribution protocol.

from qiskit import QuantumCircuit
from qiskit.converters import circuit_to_gate
from qiskit.visualization import array_to_latex
from qiskit.quantum_info import Operator
from qiskit.quantum_info import Statevector
from qiskit import transpile 
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram
from qiskit.circuit import ControlledGate
import math
import numpy as np
from qiskit.providers.fake_provider import GenericBackendV2

# Useful constants
root2 = math.sqrt(2)

denom1 = math.sqrt(4 + 2*root2)
denom2 = math.sqrt(4 - 2*root2) 


B0_transform_matrix = [ [ -1 / denom1 , (1 + root2) / denom1 ],
                        [  1 / denom2 , (root2 - 1) / denom2 ] ]
B0_transform = Operator(B0_transform_matrix) 


B1_transform_matrix = [ [  1 / denom1 , (1 + root2) / denom1 ],
                        [ -1 / denom2 , (root2 - 1) / denom2 ] ]
B1_transform = Operator(B1_transform_matrix) 


N = 100

def fiftyfifty():
    # Fill in the definition
    x = QuantumCircuit(1, 1)
    x.h(0)
    x.measure(0, 0)
    backend = BasicSimulator()
    compiled = transpile(x, backend)
    job_sim = backend.run(compiled, shots=1)
    result_sim = job_sim.result()
    counts = result_sim.get_counts(compiled)
    return int(list(counts.keys())[0]) # 0 or 1 with equal probability

def third2thirds():
    x = QuantumCircuit(1,1)
    desired_state = [1/np.sqrt(3), np.sqrt(2)/np.sqrt(3)]
    x.initialize(desired_state, 0)
    x.measure(0,0)
    backend = BasicSimulator()
    compiled = transpile(x, backend)
    job_sim = backend.run(compiled, shots=1)
    result_sim = job_sim.result()
    counts = result_sim.get_counts(compiled)
    return int(list(counts.keys())[0]) # 0 or 1 with equal probability

# The aim of the assignment is to simulate the Ekert91 key distribution protocol.

print('N =', N)

# XW, XV, ZW, ZV
arr = [[0, 0], [0, 0], [0, 0], [0, 0]]

secretKey = []

for n in range(int(9*N/2)):
    qc = QuantumCircuit(2, 3)
    qc.x(1)
    qc.h(0)
    qc.z(0)
    qc.z(1)
    qc.cx(0,1)
    #print('Initial Circuit To Get Entangled Qubits')
    #print(qc)
    
    aliceBases = ['x', 'w', 'z']
    bobBases = ['w', 'z', 'v']
    
    # Alice Chooses Basis
    aliceIndex = third2thirds()
    if aliceIndex == 1:
        aliceIndex = fiftyfifty()+1
    aliceBasis = aliceBases[aliceIndex]

    #print('Alice Index:', aliceIndex)
    #print('Alice Basis Choice:', aliceBasis)
    
    # Bob Chooses Basis
    bobIndex = third2thirds()
    if bobIndex == 1:
        bobIndex = fiftyfifty()+1
    bobBasis = bobBases[bobIndex]
    
    #print('Bob Index:', bobIndex)
    #print('Bob Basis Choice:', bobBasis)
    
    qc.barrier()

    # Attacker measures
    qc.measure(0,2)
    
    # Alice Measures
    match aliceBasis:
        case 'x':
            qc.h(0)
        case 'w':
            qc.unitary(B0_transform_matrix, 0)
            #qc.ry(-np.pi/4, 0)
        case 'z':
            pass
    qc.measure(0,0)
    
    # Bob Measures
    match bobBasis:
        case 'w':
            qc.unitary(B0_transform_matrix, 1)
            #qc.ry(-np.pi/4, 1)
        case 'z':
            pass
        case 'v':
            qc.unitary(B1_transform_matrix, 1)
            #qc.ry(-3*np.pi/4, 1)
    qc.measure(1,1)
    
    #print('Final Circuit After Basis Measurement Adjustments')
    #print(qc)
    
    # This notebook is for a simulation of the protocol without an attacker.
    backend = GenericBackendV2(num_qubits=2, control_flow=True)
    compiled = transpile(qc, backend)
    job_sim = backend.run(compiled, shots=1)
    result_sim = job_sim.result()
    counts = result_sim.get_counts(compiled)
    aliceRes = int(list(counts.keys())[0][0])
    bobRes = 1 if int(list(counts.keys())[0][1])==0 else 0
    #print(aliceRes, bobRes)
    if aliceBasis==bobBasis:
        secretKey.append(aliceRes)
    else:
        aliceRes = -1 if aliceRes==0 else 1
        bobRes = -1 if bobRes==0 else 1
        if aliceBasis=='x' and bobBasis=='w':
            arr[0][0] += aliceRes * bobRes
            arr[0][1] += 1
        elif aliceBasis=='x' and bobBasis=='v':
            arr[1][0] += aliceRes * bobRes
            arr[1][1] += 1
        elif aliceBasis=='z' and bobBasis=='w':
            arr[2][0] += aliceRes * bobRes
            arr[2][1] += 1
        elif aliceBasis=='z' and bobBasis=='v':
            arr[3][0] += aliceRes * bobRes
            arr[3][1] += 1

newArr = []
for each in arr:
    newArr.append(each[0]/each[1])
finalNum = np.abs(newArr[0] - newArr[1] + newArr[2] + newArr[3])
print('Final Res', finalNum)
print(secretKey, len(secretKey))

# This notebook is for a simulation of the protocol with an attacker, to demonstrate that the attacker can be detected.
```
